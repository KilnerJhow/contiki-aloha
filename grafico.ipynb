```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Ler os arquivos CSV
df1 = pd.read_csv('three-motes-aloha-radio-always-on.csv')
df2 = pd.read_csv('three-motes-csma.csv')

# Gerar o grfico
plt.figure(figsize=(10, 6))

# Plotar os dados do primeiro arquivo
plt.plot(df1['hops'], df1['_Eihop'], label='three-motes-aloha-radio-always-on')

# Plotar os dados do segundo arquivo
plt.plot(df2['hops'], df2['_Eihop'], label='three-motes-csma')

# Adicionar ttulo e rtulos aos eixos
plt.title('Grfico dos Dados')
plt.xlabel('hops')
plt.ylabel('_Eihop')

# Adicionar legenda
plt.legend()

# Mostrar o grfico
plt.show()
```
