```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def print_bar_graph(list_graphs):
  bar_width = 0.2

  # Gerar o gráfico
  plt.figure(figsize=(10, 6))

  n = len(list_graphs)
  for i, graph in enumerate(list_graphs):
    offset = (i - n / 2) * bar_width
    plt.bar(graph['data']['hops'] + offset, graph['data']['_Eihop'], bar_width, label=graph['label'], color=graph['color'])


  plt.title('Gráfico dos Dados')
  plt.xlabel('hops')
  plt.ylabel('_Eihop')

  # Definir os ticks do eixo x
  plt.xticks([1, 2])

  # Adicionar legenda
  plt.legend()

  # Mostrar o gráfico
  plt.show()
```
