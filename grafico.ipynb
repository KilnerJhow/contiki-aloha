```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csma_df = pd.read_csv('three-motes-csma.csv')

# Gerar o gráfico
plt.figure(figsize=(10, 6))

# Plotar os dados do primeiro arquivo
# plt.plot(df1['hops'], df1['_Eihop'], label='three-motes-aloha-radio-always-on')

# Plotar os dados do segundo arquivo
plt.bar(csma_df['hops'], csma_df['_Eihop'], label='three-motes-csma')

# Adicionar título e rótulos aos eixos
plt.title('Gráfico dos Dados')
plt.xlabel('hops')
plt.ylabel('_Eihop')

# Adicionar legenda
plt.legend()

# Mostrar o gráfico
plt.show()
```

```python
aloha_radio_always_on_df = pd.read_csv('three-motes-aloha-radio-always-on.csv')

# Gerar o gráfico
plt.figure(figsize=(10, 6))

# Plotar os dados do primeiro arquivo
plt.bar(aloha_radio_always_on_df['hops'], aloha_radio_always_on_df['_Eihop'], label='three-motes-aloha-radio-always-on')


# Adicionar título e rótulos aos eixos
plt.title('Gráfico dos Dados')
plt.xlabel('hops')
plt.ylabel('_Eihop')
plt.show()
```

```python
aloha_rdc_500_cca_active_time = pd.read_csv('three-motes-aloha-rdc-500-cca-active-time.csv')

# Gerar o gráfico
plt.figure(figsize=(10, 6))

# Plotar os dados do primeiro arquivo
plt.bar(aloha_rdc_500_cca_active_time['hops'], aloha_rdc_500_cca_active_time['_Eihop'], label='Aloha RDC 500 cca active time')

# Adicionar título e rótulos aos eixos
plt.title('Gráfico dos Dados')
plt.xlabel('hops')
plt.ylabel('_Eihop')
plt.show()
```

```python
bar_width = 0.2

# Gerar o gráfico
plt.figure(figsize=(10, 6))

# Plotar os dados do primeiro arquivo
plt.bar(aloha_radio_always_on_df['hops'] - 1.5*bar_width, aloha_radio_always_on_df['_Eihop'], bar_width, label='Aloha rádio sempre ligado', color='b')

# Plotar os dados do segundo arquivo
plt.bar(aloha_rdc_150_cca_active_time['hops'] - 0.5*bar_width, aloha_rdc_150_cca_active_time['_Eihop'], bar_width, label='Aloha RDC 150 cca active time', color='g')

# Plotar os dados do terceiro arquivo
plt.bar(csma_df['hops'] + 1.5*bar_width, csma_df['_Eihop'], bar_width, label='CSMA', color='r')

# Plotar os dados do quarto arquivo
plt.bar(aloha_rdc_500_cca_active_time['hops'] + 0.5*bar_width, aloha_rdc_500_cca_active_time['_Eihop'], bar_width, label='Aloha RDC 500 cca active time', color='y')

plt.title('Gráfico dos Dados')
plt.xlabel('hops')
plt.ylabel('_Eihop')

# Definir os ticks do eixo x
plt.xticks([1, 2])

# Adicionar legenda
plt.legend()

# Mostrar o gráfico
plt.show()
```
